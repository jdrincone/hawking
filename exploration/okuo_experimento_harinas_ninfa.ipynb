```python
# %%


def lineplot_corporativo_html(
    df: pd.DataFrame,
    x: str,
    y: str,
    color: str | None = None,
    title: str | None = None,
    yaxis_title: str = "Valor",
    xaxis_title: str | None = None,
    palette: dict | list | None = None,
    markers: bool = True,
    save_path: str = "lineplot_corporativo.html",
    frame: bool = True,
    frame_width: int = 2,
    font_family: str = "Inter, Arial, sans-serif",
    include_plotlyjs: str = "cdn",  # "cdn" para archivos livianos; usa "embed" si lo necesitas 100% offline
):
    """
    Genera un gráfico de líneas estilo corporativo (interactivo, Plotly) y lo guarda como HTML.

    Parámetros
    ----------
    df : DataFrame con los datos
    x, y : nombres de columnas para eje X e Y
    color : (opcional) columna categórica para series (similar a hue)
    title : título del gráfico
    yaxis_title, xaxis_title : títulos de ejes
    palette : dict {categoria: "#HEX"} o list ["#HEX", ...] de colores corporativos
    markers : mostrar marcadores sobre la línea
    save_path : ruta de salida .html
    frame : dibuja un recuadro negro alrededor del gráfico
    frame_width : grosor del recuadro
    font_family : fuente para títulos y ejes
    include_plotlyjs : "cdn" (liviano) o "embed" (totalmente offline)

    Retorna
    -------
    fig, save_path
    """
    # Plantilla base
    fig = go.Figure()

    # Limpieza mínima para evitar problemas en hover
    d = df.copy()

    # Resolver series
    if color is None:
        series_vals = [None]
    else:
        series_vals = pd.Series(d[color].astype("string")).fillna("Sin categoría").unique().tolist()

    # Paleta por defecto (tonos verdes + acentos)
    default_palette = ["#064E3B", "#10B981", "#0EA5E9", "#F59E0B", "#EF4444", "#6366F1", "#64748B", "#14B8A6"]

    def color_for(idx_or_key):
        if isinstance(palette, dict):
            # Si palette es dict, se espera key categórica
            return palette.get(idx_or_key, default_palette[hash(str(idx_or_key)) % len(default_palette)])
        elif isinstance(palette, list) and len(palette) > 0:
            # Si palette es lista, rotamos
            return palette[idx_or_key % len(palette)]
        else:
            # Fallback por defecto
            if isinstance(idx_or_key, int):
                return default_palette[idx_or_key % len(default_palette)]
            return default_palette[hash(str(idx_or_key)) % len(default_palette)]

    # Añadir trazas
    if color is None:
        dd = d.sort_values(x)
        fig.add_trace(
            go.Scatter(
                x=dd[x],
                y=dd[y],
                mode="lines+markers" if markers else "lines",
                name=y,
                line=dict(width=2, color=color_for(0)),
                marker=dict(size=6),
                hovertemplate=f"<b>{x}</b>: %{{x}}<br><b>{y}</b>: %{{y:.3f}}<extra></extra>",
            )
        )
    else:
        for i, val in enumerate(series_vals):
            dd = d[d[color].astype("string").fillna("Sin categoría") == str(val)].sort_values(x)
            fig.add_trace(
                go.Scatter(
                    x=dd[x],
                    y=dd[y],
                    mode="lines+markers" if markers else "lines",
                    name=str(val),
                    line=dict(width=2, color=color_for(val if isinstance(palette, dict) else i)),
                    marker=dict(size=6),
                    hovertemplate=f"<b>{color}</b>: {val}<br><b>{x}</b>: %{{x}}<br><b>{y}</b>: %{{y:.3f}}<extra></extra>",
                )
            )

    # Estética general
    fig.update_layout(
        template="plotly_white",
        title=dict(text=title, x=0.02, xanchor="left") if title else None,
        xaxis_title=xaxis_title if xaxis_title is not None else x,
        yaxis_title=yaxis_title,
        font=dict(family=font_family, size=14),
        legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1),
        margin=dict(l=60, r=40, t=60 if title else 30, b=50),
        paper_bgcolor="white",
        plot_bgcolor="white",
    )

    # Grid sutil
    fig.update_xaxes(showgrid=True, gridwidth=1, gridcolor="rgba(0,0,0,0.05)", zeroline=False)
    fig.update_yaxes(showgrid=True, gridwidth=1, gridcolor="rgba(0,0,0,0.08)", zeroline=False)

    # Recuadro en negro
    if frame:
        fig.update_layout(
            shapes=[
                dict(
                    type="rect",
                    xref="paper", yref="paper",
                    x0=0, y0=0, x1=1, y1=1,
                    line=dict(color="black", width=frame_width),
                    fillcolor="rgba(0,0,0,0)",
                    layer="above"
                )
            ]
        )

    # Guardar HTML
    return fig

```

```python


def lineplot_corporativo_subplots_html(
    df: pd.DataFrame,
    x: str,
    color: str,
    y_specs: list,  # lista de dicts: [{"y": "Aw*", "title": "...", "yaxis_title":"aW"}, ...]
    title: str | None = None,
    palette: dict | list | None = None,  # dict {categoria:"#HEX"} o list de "#HEX"
    markers: bool = True,
    frame: bool = True,
    frame_width: int = 2,
    save_path: str = "/mnt/data/panel_1x3.html",
    include_plotlyjs: str = "cdn",
    font_family: str = "Inter, Arial, sans-serif",
    horizontal_spacing: float = 0.08,
):
    """
    Crea un panel 1x3 de líneas (un subplot por métrica) con estilo corporativo y lo guarda como HTML.
    """
    assert len(y_specs) == 3, "y_specs debe tener exactamente tres elementos para un panel 1x3."

    # categorías (tratamientos) y paleta
    cats = (
        df[color].astype("string")
        .fillna("Sin categoría")
        .unique()
        .tolist()
    )
    default_palette = ["#064E3B", "#10B981", "#0EA5E9", "#F59E0B", "#EF4444", "#6366F1", "#64748B", "#14B8A6"]

    def color_for(idx_or_key):
        if isinstance(palette, dict):
            return palette.get(idx_or_key, default_palette[hash(str(idx_or_key)) % len(default_palette)])
        elif isinstance(palette, list) and len(palette) > 0:
            # idx esperado
            if isinstance(idx_or_key, int):
                return palette[idx_or_key % len(palette)]
            # key categórica
            return palette[hash(str(idx_or_key)) % len(palette)]
        else:
            if isinstance(idx_or_key, int):
                return default_palette[idx_or_key % len(default_palette)]
            return default_palette[hash(str(idx_or_key)) % len(default_palette)]

    # crear subplots
    fig = make_subplots(
        rows=1, cols=3,
        shared_xaxes=False,
        horizontal_spacing=horizontal_spacing,
        subplot_titles=[spec.get("title", spec["y"]) for spec in y_specs]
    )

    # añadir trazas por subplot
    for j, spec in enumerate(y_specs, start=1):
        ycol = spec["y"]
        # una sola vez mostramos la leyenda (en el primer subplot)
        show_legend_here = (j == 1)
        for i, cat in enumerate(cats):
            dd = df[df[color].astype("string").fillna("Sin categoría") == str(cat)].sort_values(x)
            fig.add_trace(
                go.Scatter(
                    x=dd[x],
                    y=dd[ycol],
                    mode="lines+markers" if markers else "lines",
                    name=str(cat),
                    legendgroup=str(cat),
                    showlegend=show_legend_here,
                    line=dict(width=2, color=color_for(cat if isinstance(palette, dict) else i)),
                    marker=dict(size=6),
                    hovertemplate=(
                        f"<b>{color}</b>: {cat}<br>"
                        f"<b>{x}</b>: %{{x}}<br>"
                        f"<b>{ycol}</b>: %{{y:.3f}}<extra></extra>"
                    ),
                ),
                row=1, col=j
            )

        # títulos de ejes por subplot
        fig.update_xaxes(title_text=spec.get("xaxis_title", x), showgrid=True,
                         gridwidth=1, gridcolor="rgba(0,0,0,0.05)", zeroline=False, row=1, col=j)
        fig.update_yaxes(title_text=spec.get("yaxis_title", ycol), showgrid=True,
                         gridwidth=1, gridcolor="rgba(0,0,0,0.08)", zeroline=False, row=1, col=j)

    # layout general
    fig.update_layout(
        template="plotly_white",
        title=dict(text=title, x=0.02, xanchor="left") if title else None,
        font=dict(family=font_family, size=14),
        legend=dict(orientation="h", yanchor="bottom", y=1.08, xanchor="right", x=1),
        margin=dict(l=60, r=40, t=80 if title else 60, b=50),
        paper_bgcolor="white",
        plot_bgcolor="white",
    )

    # recuadro negro alrededor de todo el panel
    if frame:
        fig.update_layout(
            shapes=[dict(
                type="rect", xref="paper", yref="paper",
                x0=0, y0=0, x1=1, y1=1,
                line=dict(color="black", width=frame_width),
                fillcolor="rgba(0,0,0,0)",
                layer="above"
            )]
        )

    return fig
```

```python

from typing import Optional
def bar_corporativo_simple(
    df: pd.DataFrame,
    cat_col: str,
    val_col: str,
    html_path: Optional[str] = None,
    *,
    orientation: str = "v",                 # "v" vertical | "h" horizontal
    title: str = "",
    x_title: Optional[str] = None,          # si no pasas, usa el nombre de la columna
    y_title: Optional[str] = None,
    value_tickformat: str = ",.2f",         # formato para el eje de valores (".0%" si son proporciones)
    textfmt: Optional[str] = None,          # por defecto usa value_tickformat
    x_tickangle: int = -30,                 # rotación etiquetas del eje X
    width: int = 900,
    height: int = 480,
):
    """
    Barra simple SIN agregación. Una fila = una barra.
    Mismo color para la misma categoría (paleta corporativa cíclica).
    """
    palette = ["#1C8074", "#666666", "#1A494C", "#94AF92", "#E6ECD8",
               "#C9C9C9", "#F6B27A", "#F5AD68", "#C76931"]

    if cat_col not in df.columns or val_col not in df.columns:
        raise ValueError("cat_col o val_col no existen en el DataFrame.")

    # Datos
    x = df[cat_col].astype(str).fillna("Sin especificar")
    y = pd.to_numeric(df[val_col], errors="coerce")

    # Colores consistentes por categoría
    cats = pd.Index(x).unique()
    color_map = {c: palette[i % len(palette)] for i, c in enumerate(cats)}
    colors = x.map(color_map)

    # Texto mostrado en cada barra
    if textfmt is None:
        textfmt = value_tickformat
    text_vals = [f"{v:{textfmt}}" if pd.notna(v) else "" for v in y]

    # Gráfico
    fig = go.Figure(go.Bar(
        x=x if orientation == "v" else None,
        y=y if orientation == "v" else None,
        orientation=orientation,
        marker=dict(color=colors, line=dict(color="rgba(0,0,0,0.25)", width=0.8)),
        text=text_vals,
        textposition="outside",
        hovertemplate=(
            f"{cat_col}: <b>%{{x}}</b><br>{val_col}: %{{y:{value_tickformat}}}<extra></extra>"
            if orientation == "v"
            else f"{cat_col}: <b>%{{y}}</b><br>{val_col}: %{{x:{value_tickformat}}}<extra></extra>"
        ),
    ))

    # Layout
    fig.update_layout(
        title=f"<b>{title}</b>" if title else None,
        font=dict(family="Inter, 'Helvetica Neue', Arial, sans-serif", size=12, color="#000"),
        paper_bgcolor="#FFFFFF", plot_bgcolor="#FFFFFF",
        width=width, height=height,
        margin=dict(l=60, r=40, t=60 if bool(title) else 30, b=60),
        bargap=0.25,
        shapes=[dict(type="rect", xref="paper", yref="paper",
                     x0=0, y0=0, x1=1, y1=1,
                     line=dict(color="#000000", width=1.0),
                     fillcolor="rgba(0,0,0,0)")]
    )

    # Títulos de ejes + formato del eje de valores
    if orientation == "v":
        fig.update_xaxes(
            title_text=x_title or cat_col, showline=True, linecolor="#000", ticks="outside",
            tickcolor="#000", tickangle=x_tickangle
        )
        fig.update_yaxes(
            title_text=y_title or val_col, tickformat=value_tickformat,
            showline=True, linecolor="#000", gridcolor="rgba(0,0,0,0.08)", rangemode="tozero"
        )
    else:  # horizontal
        fig.update_yaxes(
            title_text=y_title or cat_col, showline=True, linecolor="#000"
        )
        fig.update_xaxes(
            title_text=x_title or val_col, tickformat=value_tickformat,
            showline=True, linecolor="#000", gridcolor="rgba(0,0,0,0.08)", rangemode="tozero"
        )

    if html_path:
        fig.write_html(html_path, include_plotlyjs="cdn")
    return fig
```

```python

from typing import List, Tuple, Optional

# ---- helper: una traza de barras con colores corporativos ----
def _bar_trace_corporate(
    df: pd.DataFrame,
    cat_col: str,
    val_col: str,
    palette: Optional[List[str]] = None,
    textfmt: str = ".2f",
) -> go.Bar:
    if palette is None:
        palette = ["#1C8074", "#666666", "#1A494C", "#94AF92", "#E6ECD8",
                   "#C9C9C9", "#F6B27A", "#F5AD68", "#C76931"]

    x = df[cat_col].astype(str).fillna("Sin especificar")
    y = pd.to_numeric(df[val_col], errors="coerce")

    cats = pd.Index(x).unique()
    color_map = {c: palette[i % len(palette)] for i, c in enumerate(cats)}
    colors = x.map(color_map)

    return go.Bar(
        x=x, y=y, marker=dict(color=colors, line=dict(color="rgba(0,0,0,0.25)", width=0.8)),
        text=[f"{v:{textfmt}}" if pd.notna(v) else "" for v in y],
        textposition="outside", hovertemplate=f"{cat_col}: <b>%{{x}}</b><br>{val_col}: %{{y:{textfmt}}}<extra></extra>",
        showlegend=False
    )

# ---- función principal: grid 2x2 ----
def bar_subplots_2x2_corporate(
    datasets: List[Tuple[str, pd.DataFrame]],  # [(titulo_subplot, df), ...] (exactamente 4)
    cat_col: str = "ensayo",
    val_col: str = "resultado_log",
    *,
    title: str = "Σlog(UFC/g) por ensayo",
    x_title: str = "Ensayo",
    y_title: str = "Σlog (UFC/g)",
    value_tickformat: str = ".2f",
    x_tickangle: int = 0,
    width: int = 950,
    height: int = 700,
    shared_y: bool = True,
    write_html: Optional[str] = None,
) -> go.Figure:
    if len(datasets) != 4:
        raise ValueError("datasets debe tener exactamente 4 elementos (uno por subplot).")

    fig = make_subplots(
        rows=2, cols=2,
        subplot_titles=[t for t, _ in datasets],
        shared_yaxes=shared_y
    )

    for i, (sub_title, df_i) in enumerate(datasets):
        r = i // 2 + 1
        c = i % 2 + 1
        trace = _bar_trace_corporate(df_i, cat_col, val_col, textfmt=value_tickformat)
        fig.add_trace(trace, row=r, col=c)

        # Ejes por subplot
        fig.update_xaxes(title_text=x_title, tickangle=x_tickangle,
                         showline=True, linecolor="#000", ticks="outside", tickcolor="#000",
                         row=r, col=c)
        fig.update_yaxes(title_text=(y_title if c == 1 else None),  # solo la 1ra col para no saturar
                         tickformat=value_tickformat,
                         showline=True, linecolor="#000", gridcolor="rgba(0,0,0,0.08)",
                         rangemode="tozero", row=r, col=c)

    # Layout general
    fig.update_layout(
        title=f"<b>{title}</b>",
        font=dict(family="Inter, 'Helvetica Neue', Arial, sans-serif", size=12, color="#000"),
        paper_bgcolor="#FFFFFF", plot_bgcolor="#FFFFFF",
        width=width, height=height, bargap=0.25,
        margin=dict(l=70, r=40, t=70, b=60),
        hoverlabel=dict(bgcolor="#FFFFFF", bordercolor="#666"),
        shapes=[dict(type="rect", xref="paper", yref="paper",
                     x0=0, y0=0, x1=1, y1=1, line=dict(color="#000", width=1.2),
                     fillcolor="rgba(0,0,0,0)")]
    )

    fig.update_yaxes(range=[0, 3])   # aplica a todos los subplots

    if write_html:
        fig.write_html(write_html, include_plotlyjs="cdn")
    return fig
```

```python


# Paleta
CORP_DEEP_TEAL = "#1A494C"
CORP_TEAL      = "#1C8074"
CORP_SAGE      = "#94AF92"
CORP_GRAY      = "#6B7280"
ACCENT_WARM    =  "#C0392B"

def fig_inicial_vs_tratamientos_con_delta(df_metric: pd.DataFrame, *,
                                          titulo: str, ytitle: str,
                                          orden_trat=("Inicial","Control","T1","T2","T3")):
    """
    df_metric: columnas ['fase' ('Inicio'/'Fin'), 'trat', 'valor', 'metrica'(opcional)]
    Dibuja barras Inicio vs Fin, añade Δ vs 'Inicial' y flechas ▲/▼.
    """
    d = df_metric.copy()
    d["trat"] = pd.Categorical(d["trat"], categories=list(orden_trat), ordered=True)

    # Base (Inicial) -> si no hay 'Inicial', usa el primer valor de 'Inicio'
    base = d[(d["fase"]=="Inicio") & (d["trat"]=="Inicial")]["valor"]
    if base.empty:
        base = d.loc[d["fase"]=="Inicio","valor"]
    if base.empty:
        raise ValueError("No hay valor base de 'Inicio'.")
    base_value = float(base.iloc[0])

    # Etiqueta con Δ y color por punto
    def make_label(row):
        if row["fase"] == "Inicio":
            delta = 0.0
        else:
            delta = row["valor"] - base_value
        # Triángulos como entidades HTML para máxima compatibilidad
        arrow = "&#9650;" if delta > 1e-12 else ("&#9660;" if delta < -1e-12 else "•")
        return f"{row['valor']:.2f}<br>Δ {delta:+.2f} {arrow}"

    def label_color(row):
        if row["fase"] == "Fin":
            diff = row["valor"] - base_value
            if diff > 1e-12:
                return ACCENT_WARM
            if diff < -1e-12:
                return CORP_SAGE
        return CORP_GRAY

    d["label"] = d.apply(make_label, axis=1)
    d["label_color"] = d.apply(label_color, axis=1)

    fig = px.bar(
        d, x="trat", y="valor", color="fase",
        color_discrete_map={"Inicio": CORP_TEAL, "Fin": CORP_DEEP_TEAL},
        text="label",
        category_orders={"trat": list(orden_trat)},
        title=titulo
    )

    # Usa el text que construimos y evita recortes
    fig.update_traces(
        texttemplate="%{text}",
        textposition="outside",
        marker_line_color=CORP_GRAY,
        marker_line_width=0.5,
        cliponaxis=False
    )

    # Colorear el texto por punto
    for tr in fig.data:
        fase = tr.name
        mask = d["fase"] == fase
        tr.textfont = dict(color=d.loc[mask, "label_color"].tolist())

    # Layout corporativo
    ymax = max(d["valor"]) * 1.2
    fig.update_layout(
        paper_bgcolor="white", plot_bgcolor="white",
        font=dict(color=CORP_DEEP_TEAL),
        margin=dict(l=50, r=30, t=60, b=50),
        legend=dict(title="Fase", orientation="h", yanchor="bottom", y=1.02, xanchor="left", x=0),
        bargap=0.25, bargroupgap=0.12
    )
    fig.update_xaxes(showgrid=False, zeroline=False, showline=True, linecolor="rgba(0,0,0,0.15)")
    fig.update_yaxes(range=[0, ymax], title=ytitle, showgrid=True, gridcolor="rgba(0,0,0,0.08)",
                     zeroline=False, showline=True, linecolor="rgba(0,0,0,0.15)")

    # Recuadro
    fig.add_shape(type="rect", xref="paper", yref="paper",
                  x0=0, y0=0, x1=1, y1=1,
                  line=dict(color=CORP_GRAY, width=1),
                  fillcolor="rgba(0,0,0,0)", layer="below")
    return fig
```
