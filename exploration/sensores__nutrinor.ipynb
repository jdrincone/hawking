```python
import pandas as pd
import numpy as np
import sys
import os
sys.path.append(os.path.abspath(".."))
from core.viz import plot_line, create_subplot_grid, plot_bar, plot_statistical_strip, plot_scatter
from core.s3 import S3AssetManager
import seaborn as sns
from scipy.stats import t
import math
import matplotlib.pyplot as plt
```

```python
in_tm = pd.read_csv("../raw/temp_entrada_enfriador_nutrinor.csv")
out_tm = pd.read_csv("../raw/temp_salida_enfriador_nutrinor.csv")
air_tm = pd.read_csv("../raw/temp_entrada_aire_enfriador.csv")
```

```python
in_tm["source"] = "in"
out_tm["source"] = "out"
air_tm["source"] = "air"
tm = pd.concat([in_tm, out_tm, air_tm])

tm['date'] = pd.to_datetime(tm['date'])

tm['date_hora_str'] = tm['date'].dt.strftime('%Y-%m-%d %H')
```

```python
tm
```

```python
tm_gt = tm.groupby(["date_hora_str", "source"]).agg({"value": "mean"}).reset_index()
tm_gt
```

```python
plot_line(tm_gt, x_col="date_hora_str", y_col="value", title="Temperatura", group_col='source')
```
