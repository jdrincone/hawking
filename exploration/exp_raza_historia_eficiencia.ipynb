```python
df = pd.read_excel(
    "../raw/BD Eficiencia.xlsx", skiprows=1)
```

```python
df["FECHA"] = pd.to_datetime(df["FECHA"])
df["Mes"] = pd.to_datetime(df["FECHA"], format="%Y-%m").dt.strftime("%Y-%m")

# Número de semana ISO (1–53) y año ISO (¡puede diferir del año calendario!)
iso = df["FECHA"].dt.isocalendar()
df["Semana_ISO"] = iso.week
df["Anio_ISO"]   = iso.year

# Si quieres "YYYY-WW" en una sola columna:
df["AnioMes_ISO"] = iso.year.astype(str) + "-" + iso.week.astype(str).str.zfill(2)
```

```python
cond_rend = df["RENDIMIENTO"]<10
cond_maq = df["MAQUINA / LINEA"].isin(['2 Y 3', '4', '1', '617-620', '620-1020'])
cond_date = df["FECHA"]>"2024-01-01"
cond_tipo_proceso = df["TIPO PROCESO"].isin(["PELETIZADO", "EXTRUIDO"])

df_filter = df[(cond_rend & cond_maq & cond_date & cond_tipo_proceso)].copy()
```

```python
df_filter_comp = df_filter.groupby([ "TIPO PROCESO", "MAQUINA / LINEA", "Adición Adiflow"]).agg(
    min_date=("FECHA", "min"),
    max_date=("FECHA", "max"),
    tot_ton = ("TONELADAS", "sum"),
    tot_horas =("HORAS", "sum"),
   rendimiento_mean=('RENDIMIENTO', "mean"),
   rendimiento_median=('RENDIMIENTO', "median"),)
df_filter_comp["rendimiento_total"] = df_filter_comp["tot_ton"]/ df_filter_comp["tot_horas"]
df_filter_comp
```

```python
pt = df_filter.pivot_table(
    index=["TIPO PROCESO", "MAQUINA / LINEA"],
    columns="Adición Adiflow",          # valores típicos: 'Sí'/'No' o True/False
    values="RENDIMIENTO",
    aggfunc="mean"
).reset_index()

pt["incremento"] = (-pt["Sin Adiflow"] + pt["Adiflow"])
pt["cambio_pocentual"] = pt["incremento"]/pt["Sin Adiflow"]*100
pt
```
